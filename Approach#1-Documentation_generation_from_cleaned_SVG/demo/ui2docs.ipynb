### ü§ñ System setup


```python
system_prompt = """
You are an expert in design in different fields, you use different vector graphics tools.

You deal professionally with SVG files, you know how to create complex designs with them.

You deal professionally with UI/UX design, you know how to understand it. 

Your main jon is to analyse the UI/UX represented in the SVG format and answer relevant questions, create documentation, user stories and describe flow between frames. 

These initial inputs are SVG files that can consist of multiple layers, groups, paths, interactions, etc.

SVG contain several frame, each frame is a different UI screen. 

Each frame can contain elements as icons, logos, buttons, checkboxes, etc.

Frames can contain flows. Flows are the interactions between frames of elements of frames, such as click, scroll, etc. You should pay attention to the flow, interaction and attributes like event-type and action-type to be able to describe interactions inside the app. 
"""
```

### üñºÔ∏è Load the SVG with UI and Flows

```python
FILENAME = "music-app-flow"
svg_optimized = load_raw_svg_optimize_clean(FILENAME)
show_svg(f'./data/{FILENAME}_optimized.svg')
```

### ‚ùìWhat we want

```python
question = f"""
Here is UI/UX of a Music app in the SVG format:

{svg_optimized}

Please, generate the documentation for this app and user stories. Generate top-5 user stories.  You should pay attention to the "flow", "flows", "interaction" and attributes like "event-type", "action-type" and "destination" to be able to describe interactions inside the app. Note that using "destination" SVG describes to which element or screens leads the flow. Describe the defined interactions in this app design. Structure your answer with markdown. 
"""
```

### ‚ú® Magic

```python
chatgpt = ChatGPTWrapper(system_prompt, model="gpt-3.5-turbo-16k", temperature=0.1)
answer = chatgpt.generate(question)
```

```python
Markdown(answer)
```
